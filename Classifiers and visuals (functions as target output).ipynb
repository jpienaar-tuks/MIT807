# Imports

```python
import os, sqlite3
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score as accuracy
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from string import Template
from time import perf_counter
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

# Setting up

```python
conn = sqlite3.connect('pflacco.sqlite')
conn.row_factory = sqlite3.Row
cur = conn.cursor()

#print(conn.execute('select * from sqlite_master').fetchall())

dimensions = [2,5,10,20]
samplers = ['Latin Hypercube',
            'Random walk',
            'Hilbert curve']
feature_sets = ['disp',
                'ela_distr',
                'ela_meta',
                'ic']

# Eval budget multipliers, needs to be considered with dimensions for actual sample_size
# i.e. sample_size=178, dimensions=5, actual sample_size = 178*5=890
sample_sizes = [100, 178, 316, 562, 1000] 

# Using defaults as far as possible as per Renau paper
classifiers = {'dtree': DecisionTreeClassifier, 'knn': KNeighborsClassifier, 'rfc': RandomForestClassifier}
```

```python
def get_Xy_test_train(conn, query, params):
    df = pd.read_sql(query, conn, params=params)
    y=df.pivot_table(values='function',index='test_id', dropna=True)
    X=df.pivot_table(values='pflacco_value', index='test_id', columns='pflacco_feature', dropna=True)
    nan_rows = X.index[X.isna().any(axis=1)]
    X = X.dropna()
    scaler = StandardScaler().fit(X)
    y = y.drop(nan_rows)
    test_idx=[]
    for k, group in y.groupby('function'):
        test_idx.extend(np.random.choice(group.index,3,replace=False))
    train_idx = [i for i in y.index if i not in test_idx]
    X_test = scaler.transform(X.loc[test_idx, :])
    X_train = scaler.transform(X.loc[train_idx, :])

    y_test = y.loc[test_idx, 'function']
    y_train = y.loc[train_idx, 'function']
    return (X_test, X_train, y_test, y_train, nan_rows)

def train_classifier(classifier, X_train, y_train):
    cls = classifier().fit(X_train, y_train)
    return cls

def evaluate_classifier(classifier, X_test, y_test, nan_rows):
    pred = classifier.predict(X_test)
    acc = accuracy(y_test, pred)
    cm = confusion_matrix(y_test, pred)
    result = {'sampler': sampler,
              'feature_set': feature_set,
              'dimension': dimension,
              'sample_size': sample_size,
              'classifier': name,
              'accuracy': acc,
              'number_of_nan_rows':len(nan_rows),
              'nan_test_idx':str(nan_rows)}
    return cm, result
```

```python
# y=df.pivot_table(values='function',index='test_id', dropna=True)
# x=df.pivot_table(values='pflacco_value', index='test_id', columns='pflacco_feature', dropna=True)
# test_train_split! maybe randomly select instances?
# evaluate per feature_set

query = """
select * from features 
where
sampler = :sampler and
pflacco_feature like :feature_set and
dimensions = :dimension and
sample_n = :sample_size and
pflacco_feature != "ela_meta.quad_simple.cond" and
pflacco_feature not like "%runtime"
"""
# ela_meta.quad_simple.cond seems to cause a lot of rows to be dropped (nan and inf values)

pd.options.mode.use_inf_as_na = True

results = []
plt.tight_layout()
start = perf_counter()

for feature_set in feature_sets:
    for sample_size in [sample_sizes[0], sample_sizes[-1]]:
        for name, classifier in classifiers.items():
            fig, ax = plt.subplots(4,3,figsize=(32,18), dpi=120)
            for row, dimension in enumerate(dimensions):
                for col, sampler in enumerate(samplers):
                    ax_ = ax[row,col]
                    # Data loading, transformation and test_train splitting
                    params = {'sampler':sampler, 'feature_set':feature_set+'%', 'dimension':dimension, 'sample_size':sample_size}
                    X_test, X_train, y_test, y_train, nan_rows = get_Xy_test_train(conn, query, params)
                    
                    # Model training and calculating metrics
                    cls = train_classifier(classifier, X_train, y_train)
                    cm, result = evaluate_classifier(cls, X_test, y_test, nan_rows)
                    
                    # Prepare visuals and results list
                    #sns.heatmap(cm, ax=ax_)
                    #ax_.set_xlabel(sampler)
                    #ax_.set_ylabel(f'D={dimension}')
                    results.append(result)
            #fig.savefig(os.path.join('.','Experiment 1',f'{name} - {sample_size} - {feature_set}.png'),dpi=120)
                    
pd.DataFrame(results).to_csv(os.path.join('.','Experiment 1','Experiment 1 Accuracy scores (rfc).csv'))
print(f'That took {perf_counter()-start:2f}s')
```
