# Some function definitions
Mostly just samplers with a similar function signature, also scales output of base samplers into the $[-5,5]^d$ bounding box

```python
def hypercube_scaler(points, bounding_box):
    span = max(bounding_box)-min(bounding_box)
    offset = min (bounding_box)
    # print(f'span:{span} offset:{offset}')
    return points*span+offset

def LHS_sampler(dimensions, n, bounding_box=[-5,5]):
    LHS_sampler = qmc.LatinHypercube(dimensions)
    LHS_samples = LHS_sampler.random(n)
    LHS_samples_scaled = hypercube_scaler(LHS_samples, bounding_box)
    # pflacco 'calculate_information_content' already does nearest neighbor sorting
    # LHS_samples_ordered = greedy_nearest_neighbor(LHS_samples_scaled)
    return LHS_samples_scaled

def RW_sampler(dimensions, n, bounding_box=[-5,5]):
    walk = np.cumsum(np.random.uniform(-1,1,(n, dimensions)), axis=0)
    walk = MinMaxScaler().fit_transform(walk)
    scaled_walk = hypercube_scaler(walk, bounding_box)
    return scaled_walk

def HC_sampler(dimensions, n, bounding_box=[-5,5]):
    # Pick number of iterations such that it will generate more than 
    # the required number of samples. Also set a minimum of 3 iterations
    iterations = max(3,int(np.ceil(np.log2(n)/dimensions)))
    hilbert_curve = HilbertCurve(iterations, dimensions, -1)
    # distances = list(range(hilbert_curve.max_h+1)) # <--- SUBSAMPLE THIS
    int64_max = np.iinfo(np.int64).max
    if 2**(iterations*dimensions)+1 <= int64_max:
        distances = randomstate.choice(2**(iterations*dimensions)+1, n, replace = False, shuffle=False)
    else:
        distances = randomstate.choice(int64_max, n, replace = False, shuffle=False)
        distances = list(map(lambda x: int(x*(2**(iterations*dimensions)+1)/int64_max), distances))
    distances.sort()
    points = []
    for d in distances:
        points.append(hilbert_curve.point_from_distance(d))
    points_near_vertexes=[]
    # Adding some stochasticity
    for point in points:
        points_near_vertexes.append(np.random.normal(point,0.3))
    points_near_vertexes = np.array(points_near_vertexes)
    # Scale to unit square
    points_near_vertexes = MinMaxScaler().fit_transform(points_near_vertexes)
    # Scaling into the bounding box
    points_near_vertexes = hypercube_scaler(points_near_vertexes, bounding_box)
    return points_near_vertexes

def my_calculate_information_content(X, y):
    if name == 'Latin Hypercube':
        result=calculate_information_content(X, y)
    else:
        result=calculate_information_content(X, y, ic_sorting='none')
    return result

def step_sizes(points):
    return [np.sqrt(np.sum((points[i] - points[i+1])**2)) for i in range(len(points)-1)]

def timer(f, *args, **kwargs):
    start = perf_counter()
    res = f(*args, **kwargs)
    return (perf_counter()-start, res)
```

# Some initialisation stuff...

```python
randomstate=np.random.default_rng()
conn = sqlite3.connect(r'pflacco.sqlite')
cur = conn.cursor()
bounding_box=[-5,5]
problem_class = ProblemClass.BBOB
```

# Timing study
I'm going to break my timing study into two components:
1. Sampling from each of the samplers LHS and HC for each of the dimensions $d \in [5,10,20,30]$ for each of the sample_sizes $n \in [100,316,1000]$, calculating the y-values for all of the BBOB problems, 1st instance and finally calculating the information content features, while keeping track of time spent on a) sample generation and b) feature calculation
2. Running multiple repetitions of the above, but only in $d=5$ to see if HC features generated from the HC sampler converges to similar values to the LHS. I also want to examine some summary statistics of the step sizes generated by both the HC and LHS samplers (mean, std dev). I also expect the LHS stepsize sequence to be skewed and want to see if a Gaussian Mixture model can be fit to it...

## Data generation: Part 1

```python
cur.execute('create table if not exists timings (test_id integer, \
                                                function integer, \
                                                dimensions integer, \
                                                sample_n integer, \
                                                sampler text, \
                                                feature text, \
                                                value float)')
test_id=0
samplers = {'Latin Hypercube': (LHS_sampler, 'nn'), 'Hilbert Curve': (HC_sampler, 'none')}
warnings.filterwarnings('ignore')
test_id_done = cur.execute('select max(test_id) from timings').fetchone()[0]
for dimension in [5,10,20,30]:
    for sample_size in [100,316,1000]:
        for name, (sampler, ic_sorting) in samplers.items():
            for function in problem_class.problems:
                test_id+=1
                if test_id <= test_id_done:
                    continue
                test_dict={'dimensions':dimension, 
                           'sample_n': sample_size, 
                           'function': function, 
                           'test_id': test_id, 
                           'sampler':name}
                results={}
                results['t.generation'], X = timer(sampler, dimensions=dimension, n=sample_size*dimension)
                f = get_problem(function, 1, dimension, problem_class)
                y=np.array(list(map(f,X)))
                results['t.ic.calculate'], result = timer(calculate_information_content, X, y, ic_sorting=ic_sorting)
                results.update(result)
                db_list=[]
                for k, v in results.items():
                    db_list.append((test_id, function, dimension, sample_size, name, k, v))
                cur.executemany('insert into timings (test_id, function, dimensions, sample_n, sampler, feature, value) \
                values (?,?,?,?,?,?,?)', db_list)
                conn.commit()
```

## Data generation: Part 2

```python
from sklearn.mixture import GaussianMixture
from scipy.stats import skewtest

cur.execute('create table if not exists repetitions (test_id integer, \
                                                    repetition integer, \
                                                    function integer, \
                                                    sample_n integer, \
                                                    sampler text, \
                                                    feature text, \
                                                    value float)')
test_id=0
samplers = {'Latin Hypercube': (LHS_sampler, 'nn'), 'Hilbert Curve': (HC_sampler, 'none')}
warnings.filterwarnings('ignore')
test_id_done = cur.execute('select max(test_id) from timings').fetchone()[0]
dimension=5
gmm = GaussianMixture(2)
for repetition in range(30):
    for sample_size in [100,1000]:
        for name, (sampler, ic_sorting) in samplers.items():
            for function in problem_class.problems:
                test_id+=1
                if test_id <= test_id_done:
                    continue
                test_dict={'sample_n': sample_size, 
                           'function': function, 
                           'test_id': test_id, 
                           'sampler':name}
                results={}
                results['t.generation'], X = timer(sampler, dimensions=dimension, n=sample_size*dimension)
                f = get_problem(function, 1, dimension, problem_class)
                y=np.array(list(map(f,X)))
                results['t.ic.calculate'], (step_sizes, result) = timer(calculate_information_content, X, y, ic_sorting=ic_sorting, return_step_sizes=True)
                results.update(result)
                
                # Collecting some summary statistics on the stepsizes as well as the results of fitting a Gaussian Mixture model
                # to the step size data
                step_sizes = np.array(step_sizes)
                gmm.fit(step_sizes.reshape(-1,1))
                s, p = skewtest(step_sizes)
                step_size_stats= {'sss.mean':step_sizes.mean(),
                                  'sss.std':step_sizes.std(),
                                  'sss.skew': s,
                                  'sss.skew_p': p,
                                  'sss.weight.0': gmm.weights_.squeeze()[0],
                                  'sss.weight.1': gmm.weights_.squeeze()[1],
                                  'sss.mean.0': gmm.means_.squeeze()[0],
                                  'sss.mean.1': gmm.means_.squeeze()[1],
                                  'sss.cov.0': gmm.covariances_.squeeze()[0],
                                  'sss.cov.1': gmm.covariances_.squeeze()[1]
                                 }
                results.update(step_size_stats)                                  
                
                db_list=[]
                for k, v in results.items():
                    db_list.append((test_id, function, repetition, sample_size, name, k, v))
                cur.executemany('insert into repetitions (test_id, function, repetition, sample_n, sampler, feature, value) \
                values (?,?,?,?,?,?,?)', db_list)
                conn.commit()
```

## Visualisations: Part 1

```python
title_dict={'t.generation':'Sample generation', 't.ic.calculate':'Calculation of information content features'}
df = pd.read_sql('select * from timings where feature like "t.%"',conn)
df['actual_n']=df['dimensions']*df['sample_n']
df['titles']=df['feature'].map(title_dict)
g=sns.lmplot(df, x='actual_n', y='value', hue='sampler', col='titles', facet_kws={'sharey':False}, order=2)
g.set_axis_labels('Sample size', 'Time (s)')
g.set_titles(col_template='{col_name}')
plt.show()
```

### Curious to see some regressions:

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(2, include_bias=True)
linear = LinearRegression(fit_intercept=False, positive=True)
results=dict()
scores=dict()
for gname, gdf in df.groupby(['sampler','feature']):
    X=gdf.loc[:,['dimensions','actual_n']]
    y=gdf['value']
    X=poly.fit_transform(X)
    linear.fit(X,y)
    results[gname]=linear.coef_
    scores[gname]=linear.score(X,y)
out_features = poly.get_feature_names_out()
for gname in results.keys():
    print('{gname} - Score: {score}\nCoefficients: \n{coefs}'.format(gname=gname, 
                                                                     score=scores[gname],
                                                                     coefs=''.join(["{0}: {1:g},\n".format(f, i) for f, i in zip(out_features,results[gname])])))
```
